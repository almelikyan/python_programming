## 2. Индексы и срезы <a class="anchor" id="chapter2"></a>

__Индексы__ позволяют получать доступ к элементам списка, а __срезы__ — извлекать подмножества элементов.

```python
Image("pic9.png", width="450")
```

__Индексация__ начинается с нуля: первый элемент списка имеет индекс 0, второй — 1, и так далее. Можно использовать отрицательные индексы для доступа к элементам списка с конца: последний элемент имеет индекс -1, предпоследний — -2 и так далее. 

__Срезы__ позволяют извлекать подсписки из основного списка. 

Синтаксис: `list[start:stop:step]`

- `start` — начальный индекс (включительно). Если отсутствует, то срез начинается с 0;
- `stop` — конечный индекс (не включительно!). Если отсутствует, то срез идёт до конца списка;
- `step` — шаг, с которым извлекаются элементы. По умолчанию равен 1 (каждый элемент). Значение 2 - каждый второй элемент.

```python
num_list = [34, 21, -35, 98, 0, 85, -45, 65]
```

__Индексы__

```python
num_list[0]  # первый элемент
```

```python
num_list[2]  # третий элемент
```

```python
num_list[-1]  # последний элемент
```

```python
num_list[-3]  # третий элемент с конца
```

__Срезы__

```python
num_list[0:]  # весь список
```

```python
num_list[2:-1]  # с третьего до предпоследнего
```

```python
num_list[3:]  # с четвертого до последнего
```

```python
num_list[:5]  # первые пять
```

```python
num_list[::2]  # начиная с первого элемента через один
```

```python
num_list[1:6:3]  # со второго по шестой элемент через три
```

### 3.3. Замена значений в  списке<a id="section_3_3"></a>

__Замена одного значения__

```python
print("Исходный список:", faculties)
faculties[1] = "Chemistry"  # присваиваем новое значение 2-му элементу списка, старое значение удаляется
print("Новый список:", faculties)
```

__Замена нескольких значений__

Для замены нескольких значений формируется срез из старых значений, который заменяется на список из новых значений.

```python
my_numbers = [5, 8, 9, 10, 12, 6, 18]
print("Исходный список:", my_numbers)
print("Заменяемый срез:", my_numbers[1:3])  # это срез из 2-го и 3-го значений
my_numbers[1:3]= [22, 99]  # заменяем его на список из 2 других элементов (их могло быть и больше 2)
print("Новый список:", my_numbers)
```

### 3.4. Добавление значений в список<a id="section_3_4"></a>

__Добавление через пустой срез__

Для добавления новых значений в список формируется пустой срез на место которого вставляются новые значения.

```python
my_numbers = [5, 8, 9, 10, 12, 6, 18]
print("Исходный список:", my_numbers)
print("Пустой срез:", my_numbers[1:1]) # это пустой срез (2-ая позиция в списке)

my_numbers[1:1] = [99, 77, 55]
print("Новый список:", my_numbers)
```

__Добавление элемента в конец списка__

Метод `append()` используется для добавления одного элемента в конец списка.

```python
my_numbers = [5, 8, 9, 10, 12, 6, 18]
my_numbers.append(20)  # добавляем 20 в конец списка
my_numbers
```

__Добавление элемента на указанную позицию в списке__

Метод `insert()` используется для вставки элемента в список на указанную позицию. 

Синтаксис: `list.insert(index, element)`

- `index` — позиция (индекс), куда будет вставлен элемент;

- `element` — элемент для вставки.

```python
my_numbers = [5, 8, 9, 10, 12, 6, 18]
my_numbers.insert(2, "new_element")  # добавляет элемент на 3-ую позицию
my_numbers
```

### 3.5. Удаление значений из списка<a id="section_3_5"></a>

__Удаление с помощью оператора `del()`__

Оператор `del()` позволяет удалять элементы из списка как по индексу, так и по срезу.

```python
my_numbers = [5, 8, 9, 10, 12, 6, 18]
print("Исходный список:", my_numbers)
del(my_numbers[2]) # удалим 3-ий элемент списка
print("Список без 3-го элемента:", my_numbers)
del(my_numbers[0:2]) # удалим первые 2 элемента в ранее измененном списке
print("Список без первых 2-х элементов:", my_numbers)
```

__Удаление с помощью метода `remove()`__

Метод `remove()` используется для удаления первого вхождения указанного элемента из списка. 

```python
names_list = ['Анна', 'Олег', 'Ирина', 'Василий', 'Егор', 'Ирина', 'Петр']
print('Исходный список:', names_list)
names_list.remove('Ирина')
print('Измененный список:', names_list)
```

__Удаление с помощью метода `pop()`__

Метод `pop()` удаляет элемент списка по указанному индексу и возвращает удалённое значение.

```python
names_list = ['Анна', 'Олег', 'Ирина', 'Василий', 'Егор', 'Ирина', 'Петр']
print('Исходный список:', names_list)
names_list.pop(1)  # удаляем второй элемент
print('Измененный список:', names_list)
```

### 3.12. Сортировка списка<a id="section_3_12"></a>

Метод `sort()` сортирует элементы списка, изменяя его.

Синтаксис: `list.sort(key=None, reverse=False)`

- `key` - функция, возвращающая значение для сравнения (опционально)

- `reverse` - флаг обратной сортировки (по умолчанию False - сортировка по возрастанию)

```python
names_list = ['Илья', 'Анна', 'Олег', 'Ирина', 'Василий', 'Егор', 'Ирина', 'Петр']
names_list.sort()  # сортировка по алфавиту
names_list
```

```python
my_numbers = [5, 8, 9, 10, 12, 6, 18, 1]
my_numbers.sort()  # сортировка по возрастанию
my_numbers
```

```python
names_list = ['Илья', 'Анна', 'Олег', 'Ирина', 'Василий', 'Егор', 'Ирина', 'Петр']
names_list.sort(key=len)  # сортировка по числу знаков в элементе (по умолчанию по возрастанию)
names_list
```

```python
names_list = ['Илья', 'Анна', 'Олег', 'Ирина', 'Василий', 'Егор', 'Ирина', 'Петр']
names_list.sort(key=len, reverse=True)  # сортировка по числу знаков в элементе по убыванию
names_list
```

Функция `sorted()` — это встроенная функция, которая возвращает новый отсортированный список из любого итерируемого объекта. В отличие от метода списка sort(), sorted() не изменяет исходный объект.

Синтаксис: `sorted(iterable, key=None, reverse=False)`

- `iterable` - любой итерируемый объект (список, кортеж, строка и т.д.);

- `key` - функция для определения порядка сортировки (опционально);

- `reverse` - флаг обратной сортировки (по умолчанию False).

```python
names_list = ['Илья', 'Анна', 'Олег', 'Ирина', 'Василий', 'Егор', 'Ирина', 'Петр']
sorted(names_list)  # сортировка списка по алфавиту
```

```python
names_list = ['Илья', 'Анна', 'Олег', 'Ирина', 'Василий', 'Егор', 'Ирина', 'Петр']
sorted(names_list, reverse=True)  # сортировка списка в обратной алфавитном порядке
```

```python
my_tuple = ('Илья', 'Анна', 'Олег', 'Ирина', 'Василий', 'Егор', 'Ирина', 'Петр')
sorted(my_tuple)  # сортировка кортежа, возвращает список
```

```python
word = "программирование"
sorted_chars = sorted(word)  # сортировка строки, возвращает список
print(sorted_chars)  
```

### 3.18. Сравнение списков<a id="section_3_18"></a>

Сравнение списков происходит лексикографически (поэлементно), аналогично строкам. Элементы должны быть сравнимыми. Сначала сравниваются первые элементы, если равны - вторые и так далее. Если все элементы равны, но один список короче — он считается меньшим.

__Равенство  (`==, !=`)__

Списки равны, если все элементы совпадают по порядку и значению.

```python
[1, 2, 3] == [1, 2, 3]
```

```python
[1, 2, 3] == [1, 2, 4]
```

```python
[1, 2, 3] != [1, 2, 4]
```

__Сравнение порядка (`<, >, <=, >=`)__

```python
[1, 2] < [1, 2, 3]  # второй список длиннее
```

```python
[1, 3] < [1, 2, 3]  # 3 > 2 на второй позиции
```

```python
[1, 1] < [1, 2]  # 1 < 2 на второй позиции
```

```python
[4] > [3, 5]  # 4 > 3 на первой позиции
```

```python
[1, 2] <= [1, 2]  # списки равны
```

```python
[] < [1]  # пустой меньше непустого
```

## 4. Вложенные списки <a class="anchor" id="chapter4"></a>

Элементом списка может быть другой список.

- `matrix[i]` — i-й список (ряд).

- `matrix[i][j]` — j-й элемент в i-м списке.

```python
matrix
```

```python
matrix[0]  # первый ряд
```

```python
matrix[0][1]  # второй элемент первого ряда
```

```python
matrix[2][0]  # первый элемент третьего ряда
```

```python
matrix[1] = [0, 0, 0]  # заменить вторую строку на нули
```

```python
matrix
```

```python
employees[1:3]  # выводим данные о 2-м и 3-м сотрудниках
```

```python
employees[2][0]  # выводим имя 3-го сотрудника
```

```python
employees[1][1][0]  # выводим 1-й отдел 2-го сотрудника
```

## 5. Кортежи (tuple) <a class="anchor" id="chapter5"></a>

**Кортеж** (tuple) — это упорядоченная и неизменяемая коллекция элементов. По своей структуре похожи на списки, но обладают тремя ключевыми свойствами:

- Неизменяемость — после создания кортежа невозможно добавить, удалить или изменить его элементы. 
- Упорядоченность — элементы располагаются в определённом порядке, и к ним можно обратиться по индексу.
- Гибкость содержимого — элементами кортежа могут быть объекты различных типов: числа, строки, списки, другие кортежи и другие.

# Задание 1


Напишите программу, которая запрашивает текстовую строку, в которой между словами может быть любое число пробелов, затем выводит её не более, чем с одним пробелом между словами.

_Пример:_

In: `Высшая                               школа      экономики`

Out: `Высшая школа экономики`


```python
# Здесь запишите решение задания 1 ☆*:.｡.o(≧▽≦)o.｡.:*☆
s = input().split()
print(' '.join(s))
```

# Задание 4

_Пример:_

_In:_  
_`Один Два Три Четыре`_
    
_`3`_

_Out: `Три`_

# Задание 5

# Задание 8

Напишите программу, которая получает на вход список и преобразует его так, что последние 3  элемента становятся первым, а остальные сдвигаются на 3 позиции вправо.

_Пример:_

_In: `[1, 2, 3, 4, 5]`_
      
_Out: `[3, 4, 5, 1, 2]`_

```python
# Здесь запишите решение задания 8 ʕ•́ᴥ•̀ʔっ
lst = [1, 2, 3, 4, 5, 6, 7]
result = lst[-3:] + lst[:-3]
print(result)
```
