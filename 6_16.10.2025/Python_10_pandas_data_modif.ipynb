```python
titanic[~((titanic['Sex'] == 'male') | (titanic['Age'] > 30))]  # ни одно из условий не выполняется
```

```python
# Отобрать DataFrame по полю "Age" (между 30 и 40 включительно)
titanic[titanic['Age'].between(30, 40)]
```

```python
titanic[titanic['Age'].isin([10, 20, 30])]  # пассажиры в возрасте 10, 20 или 30 лет
```

Заменим пропущенные значения возраста медианой.

```python
titanic = pd.read_csv('titanic.csv', index_col=0)
titanic['Age_median'] = titanic['Age'].fillna(titanic['Age'].median())
titanic.head(7)
```

Заменим пропущенные значения возраста медианой отдельно для мужчин и женщин

```python
titanic = pd.read_csv('titanic.csv', index_col=0)
titanic['Age_median'] = titanic['Age'].fillna(titanic.groupby('Sex')['Age'].transform("median"))
titanic.head(20)
```

4.5. Создание нового столбца на основе существующих

```python
titanic["Name_len"] = titanic["Name"].apply(len)
titanic
# будет создана новая колонка "Name_len", отражающая число знаков в столбце "Name"
```

```python
titanic["Name_upper"] = titanic["Name"].apply(str.upper)
titanic
# будет создана новая колонка "Name_upper", в которой имя пассажира записано прописными буквами
```

```python
titanic["Name"] = titanic["Name"].apply(str.upper)
titanic
# в этом случае создаётся не новая колонка, а меняется существующая
```

```python
# новый столбец, показывающий сколько всего родственников ехало с пассажиром (сумма столбцов SibSp и Parch)
titanic = pd.read_csv('titanic.csv', index_col=0)
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch']
titanic.head(5)
```

```python
# создаём булеву переменную, принимающую значение True, если пассажир плыл один, False - если не один
titanic['Alone'] = titanic['FamilySize']==0
titanic.head()
```

```python
# создаём бинарную переменную, принимающую значение 1, если пассажир плыл один, 0 - если не один
titanic['Alone2'] = titanic['FamilySize'].apply(lambda x: 1 if x==0 else 0)
titanic.head()
```

```python
# создаём бинарную переменную, принимающую значение 1, если пассажир был совершеннолетний, 0 - если нет
titanic['Adult'] = titanic['Age'].apply(lambda x: 1 if x>=18 else 0)
titanic.head(20)
```

```python
# создаём функцию
def get_age_gr(x):
    if x < 21:
        return 'young'
    elif x >=21 and x<60:
        return 'adult'
    elif x>=60:
        return 'old'
# не используем else, чтобы Nan не вошли в группу 3
```

```python
# создаём категориальную переменную на основе количественной 
titanic['Age_groups'] = titanic['Age'].apply(get_age_gr)
titanic.head(20)
```

```python
titanic['Age_groups'].value_counts()
```

```python
titanic.Age_groups.dtypes  # сейчас тип столбца object, текстовая переменная
```

Для работы с категориальными переменными иногда требуется их перекодировать в числовые, с сохранением текстовых меток, привязанных к соответствующим числам. Это можно сделать с помощью метода Categorical.

```python
titanic['Age_groups'] = pd.Categorical(titanic['Age_groups'])
```

```python
titanic['Age_groups'].cat.codes.head(34)  # каждому текстовому значению соответствует числовой код
```

```python
titanic['Age_groups'].cat.categories  # выводим список категорий
```

## 6. Ранжирование 

```python
titanic = pd.read_csv('titanic.csv', index_col=0)
titanic['Age_rank_1'] = titanic['Age'].rank()
titanic.sort_values(by='Age').head(15) 
# самый юный пассажир получает ранг 1
# в случае связанных рангов рассчитывается средний ранг
```

```python
titanic['Age_rank_2'] = titanic['Age'].rank(method='max') 
titanic.sort_values(by='Age').head(15)
# самый юный пассажир получает ранг 1
# в случае связанных рангов рассчитывается максимальный ранг
```

```python
titanic['Age_rank_3'] = titanic['Age'].rank(method='min') 
titanic.sort_values(by='Age').head(15)
# самый юный пассажир получает ранг 1
# в случае связанных рангов рассчитывается минимальный ранг
```

```python
titanic['Age_rank_4'] = titanic['Age'].rank(method='min', na_option='bottom') 
titanic.head(19)
# самый юный пассажир получает ранг 1
# в случае связанных рангов рассчитывается минимальный ранг
# если возраст не известен, наблюдение оказывается в конце рейтинга
```

```python
titanic['Age_rank_5'] = titanic['Age'].rank(ascending=False) 
titanic.sort_values(by='Age', ascending=False).head(15)
# самый пожилой пассажир получает ранг 1
```
